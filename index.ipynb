### Q6: Write a function to find the closest brewery to any other given brewery

```python
def find_closest_brewery(target):
    target_brew = [brewery for brewery in breweries100 if brewery['name']==target]
    target_coord = (target_brew[0]['latitude'], target_brew[0]['longitude'])
    all_coord = []
    distances = []

    for brewery in breweries100:
        if brewery['latitude'] == None or brewery['longitude'] == None:
            pass
        else:
            all_coord.append({'name':brewery['name'], 'coordinates': (brewery['latitude'], brewery['longitude'])})

    for coord in all_coord:
        distances.append({'name': coord['name'], 'distance': geopy.distance.vincenty(target_coord, coord['coordinates']).km})

    sorted_distances = sorted(distances, key = lambda i: i['distance']) 
    if sorted_distances[0]['distance'] == 0:
        return sorted_distances[1]['name']
    else:
        return sorted_distances[0]['name']
```

##### 2. We've queried the API once, but it could get annoying to retype the url over and over again, create a function `make_api_request` that enables you to query the API.


 The parameters for the function should be:
* lat (float) : latitude
* lng (float) : longitude
* date (string): Date in the format YYYY-MM
    * default value = `None`
    
And it should return a json object of 

for more information on default values check out http://blog.thedigitalcatonline.com/blog/2015/02/11/default-arguments-in-python/

```python
def make_api_request(lat, lng, date=None):
    import requests
    import json
    url = 'https://data.police.uk/api/crimes-street/'
    url = url+'/all-crime?lat='+str(lat)+'&lng='+str(lng)+'&date='+str(date)
    r = requests.get(url)
    return r.json()
    
```

##### 3. Write a function `categories_of_crime` that will determine the count of each type of crime for a given latitude and longitude. This is labelled as 'category' in the records. Your function should call the `make_api_request` function you created.

The parameters for the function should be:

* lat (float) : latitude
* lng (float) : longitude
* date (str) default = None

The function should return:
* a dictionary with the count of each type of crime



Once you've created the function, try it with these locations
* lat, lng of 51.5017861,-0.1432319   (Buckingham Palace)
* lat, lng of 53.480161, -2.245163     (Manchester)

```python
def categories_of_crime(lat, lng, date):
    from collections import Counter
    categories_of_crimes: []
    crimes = make_api_request(lat,lng,date)
    for crime in crimes:
        categories_of_crimes.append(crime['category'])
        result = Counter(categories_of_crimes)
    return result
```
